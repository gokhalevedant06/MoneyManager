```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
plt.figure(figsize=(17, 5))
```

```python
df = pd.read_csv('SensexHistoricalData.csv')
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.head()
```

```python
df['test'] = df.index
df
```

```python
df.dtypes
```

```python
df.set_index(df.Date, inplace=True)
df.head()
```

```python
df.drop(['Date', 'Volume', 'Currency'], axis=1, inplace=True)
df.head()
```

```python
df_open = df.filter(["Open"])
plt.plot(df_open)
```

```python
from statsmodels.tsa.stattools import adfuller
adf, pvalue, usedlag_, nobs_, crtical_values_, icbest_ = adfuller(df_open)
pvalue
```

```python
df['Year'] = [d.year for d in df_open.index]
df['Month'] = [d.strftime("%b") for d in df_open.index]
df.head()
```
